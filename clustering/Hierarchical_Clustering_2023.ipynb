## Hierarchical Clustering
![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

# Generamos datos aleatorios
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Ejecutamos el algoritmo de clustering jerárquico
clustering = AgglomerativeClustering(n_clusters=4)
clustering.fit(X)

# Visualizamos los resultados
plt.scatter(X[:,0], X[:,1], c=clustering.labels_, cmap='rainbow')
plt.show()
```

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)
### Ahora probamos con el Iris dataset

![divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

```python
from sklearn import datasets
from random import randint

iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target


# El dataset iris tiene 3 features
nro_clusters = 3


# Ejecutamos el algoritmo de clustering jerárquico
clustering = AgglomerativeClustering(n_clusters=nro_clusters)
clustering.fit(X_iris)

# Visualizamos los resultados
plt.scatter(X_iris[:,0], X_iris[:,1], c=clustering.labels_, cmap='rainbow')
plt.show()
```
